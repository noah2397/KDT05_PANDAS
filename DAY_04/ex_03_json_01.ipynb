- json 중첩 데이터의 다중 반복문 => 단일 리스트 형태 처리 : 평탄화/정규화

```python
pd.json_normalize(oneData, record_path="laureates")
```

```python
pd.json_normalize(oneData, record_path="laureates", meta=["year","category"])
```

```python
# 아래 코드를 실행하기 전에, 모든 딕셔너리에 대해 record_path에 해당하는 키 생성하기
for data in nobelDF.prizes:
    print(type(data), data.keys(), end="\n\n")
    # 해당 dict에 laureates 키가 없다면 추가 
    data.setdefault('laureates',[])
```

```python
# 컬럼 읽기 => DF객체변수명[컬럼명],DF객체변수명.컬럼명
pd.json_normalize(nobelDF.prizes, record_path='laureates') 
# -> laureates에 대한 에러가 뜬다면 없는 애가 없다는 것이다
# -> 모든행에 record_path에 해당하는 키(key)가 존재해야 함!
```

```python
nobleDF2=pd.json_normalize(nobelDF.prizes, record_path='laureates',meta=['year','category'])
nobleDF2.info()
```

- (7) JSON의 다양한 방식의 DataFrame 저장 => orient 파라미터

```python
# 딕셔너리 => 키 : 밸류
# 첫번째 케이스 : 키가 없는 상태
json_data=''' 
[
{"id" : "1","name":"sravan","age":22},
{"id" : "2","name":"harsha","age":22},
{"id" : "3","name":"deepika","age":21},
{"id" : "4","name":"jyothika","age":23}
]
'''
```

- orient = 'record' 설정 => JSON 파일 데이터를 행 단위로 읽어서 DataFrame에 저장

```python
df=pd.read_json(json_data, orient = 'record') #json객체가 여러 개 있는 배열로 판단하고, 한 행씩 읽어들인다
# 한 json객체를 레코드(행)로 본다
# 한 중괄호 덩어리를 json객체라 본다
print(df)
```

```python
# 키가 하나 붙어있는 상태
# 키 : 밸류, 키: 밸류....형식이다
# 키는 컬럼명으로 두고, 밸류는 레코드(행)로 본다
# 아까의 json_data에서는 키값이 없어서 read_json했을 때 인덱스값이 안들어갔었다
# 딕셔너리 안에 딕셔너리가 있는 형태일 때, orient='index'나 'columns'를 쓸 수 있다!
json_data2=''' 
{
    "Student1" : {"id" : "1","name":"sravan","age":22},
    "Student2" : {"id" : "2","name":"harsha","age":22},
    "Student3" : {"id" : "3","name":"deepika","age":21},
    "Student4" : {"id" : "4","name":"jyothika","age":23}
}
'''
```

```python
df=pd.read_json(json_data2, orient='index') #, orient="
print(df)
```

```python
df=pd.read_json(json_data2, orient='columns') #, orient="records")
print(df)
```

```python
df=pd.read_json(json_data2, orient='columns').T
print(df)
```

```python
# JSON Format # 키와 밸류 쌍이 없고, 오직 값밖에 없다
json_data3 = '''
[
    [ "1", "sravan",22 ],
    [ "2", "harsha",22 ],
    [ "3", "deepika",21 ],
    [ "4", "jyothika",23 ]
]
'''
```

```python
df=pd.read_json(json_data3, orient='values') # 단지 값만 들어있다고 표시
df
```
