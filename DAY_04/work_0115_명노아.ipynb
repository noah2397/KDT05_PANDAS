```python
# 1번 문제
import pandas as pd # 판다스 모듈 준비
import numpy as np # 결측치(NaN) 사용할 넘피 모듈 준비
```

```python
pikachu = pd.read_excel("./남북한발전전력량.xlsx") # 엑셀 파일 읽어오기
```

```python
pikachu.loc[9]="-" # 행 추가 ( 근데 왜 열도 하나 추가되나요....? )
```

```python
data = pikachu[[f"{str(num)}" for num in range(1990,2017)]] # 년도별 에너지 수치 데이터만 따로 저장 ["1990","1991",...,"2015","2016"]
for col in data.columns:  # 컬럼명을 반복하면서
    for idx, value in enumerate(data[col]): # enumerate로 idx와 value값을 뽑아옴
        if value=='-' : # 해당 값이 "-"" 이면 수치 데이터 상 0이라고 표현 
            data[col][idx]='0'
```

```python
for col in data.columns:
    data[col]=data[col].astype("int64") # 수치 데이터로 모두 변환을 했으니 int형으로 Series 형변환시킴
    
data=data.T # sum은 세로로 더해주니 열과 행을 뒤집어줌
result_sum = list(data.sum()) # pikachu DataFrame에 집어넣을 것이므로 list변환
result_sum
```

```python
pikachu.loc[9,"발전 전력별"]="신재생"  # 해당 부분에 값 넣기
pikachu.loc[9,"전력량 (억㎾h)"]=np.nan # 해당 부분에 값 넣기
pikachu.rename(columns={1:"총계"}, inplace=True) # 왜인지 모르겠지만 추가된 열 이름 바꾸고, inplace로 저장하기
pikachu["총계"]=result_sum # 새롭게 생겨난 열에 합산 결과를 담은 리스트 입력
pikachu # 추가
```

```python
#2번 문제
import pandas as pd # 판다스 모듈 준비
```

```python
move=pd.read_excel("시도별 전출입 인구수.xlsx") # 엑셀 데이터 읽어들이기
```

```python
data = move[[f"{str(num)}" for num in range(1970,2017)]] # 년도별 에너지 수치 데이터만 따로 저장 ["1970","1971",...,"2016","2017"]
data=data.drop([0])
print(data["1970"].value_counts().count())
for col in data.columns:  # 컬럼명을 반복하면서
    data[col] = data[col].apply(lambda x: '0' if x == '-' else x) # 이 코드로 겨우겨우 실행 됌...짜증남!!!!!
    #print(col)
    #for idx, value in enumerate(data[col]): # enumerate로 idx와 value값을 뽑아옴
    #    if value=='-' : # 해당 값이 "-" 이면 수치 데이터 상 0이라고 표현 
    #        data[col][idx]='0' <- 이 코드는 왜안되는 거지....데이터 손실이 일어남 145->133
```

```python
for col in data.columns:
    data[col]=data[col].astype("int64") # 수치 데이터로 모두 변환을 했으니 int형으로 Series 형변환시킴
data=data.T
result=list(data.sum()) # 각 가로줄 별로 합산함 -> 지역에서 지역별로 1970~2017년 동안 이동한 인구의 총 합계를 계산함
```

```python
move=pd.read_excel("시도별 전출입 인구수.xlsx", usecols=['전출지별','전입지별']) # 파일을 새로 불러들임(근데 생각해보니 필요없었음)
```

```python
move = move.drop([0]) # result(합산결과) 와 길이를 맞추기 위해서 move의 데이터 한 행 삭제
move["총 이동인구 수"]=result # "총 이동인구 수" 열을 생성하고 합산 결과를 집어넣음
# move.reindex([i for i in range(0,323)], axis='index') <- 모르겠다
```

```python
local=["전국","서울","부산","대구","인천","광주","대전","울산","세종특별자치시","경기도","강원도","충청북도","충청남도","전라북도","전라남도","경상북도","경상남도","제주특별자치도"] # 총 합산 결과의 부분과 같이 합쳐져서 Dict로 만들기 위해 지역 리스트 생성
len(local) # 지역의 종류는 총 18가지임
```

```python
move_to_seoul=[f'{move["총 이동인구 수"][i]}' for i in range(2, 325, 18)] # 서울에 이동한 지역별 사람들 수를 comprehension으로 만들기
move_to_seoul_zip=dict(zip(move_to_seoul, local)) # move_to_seoul과 local 리스트를 합쳐서 딕셔너리로 생성

# key lambda를 사용하여 "전국"을 제외한 배열 중에서 두번째로 높은 값을 뽑아냄(30분 걸림 아오!!!!!!)
first_max = max(list(map(int, move_to_seoul_zip.keys())), key=lambda x: -1 if x==max(list(map(int, move_to_seoul_zip.keys()))) else x)

# 딕셔너리의 키 값을 조회해서 해당 지역을 출력함
print(f"1970~2016년 동안 서울로 제일 많이 간 사람들의 지역 : {move_to_seoul_zip[str(first_max)]}")
```

```python
move_to_seoul=[f'{move["총 이동인구 수"][i]}' for i in range(18, 325, 18)]# 서울에 이동한 지역별 사람들 수를 comprehension으로 만들기
move_to_seoul_zip=dict(zip(move_to_seoul, local)) # move_to_seoul과 local 리스트를 합쳐서 딕셔너리로 생성

# key lambda를 사용하여 "전국"을 제외한 배열 중에서 두번째로 높은 값을 뽑아냄
first_max = max(list(map(int, move_to_seoul_zip.keys())), key=lambda x: -1 if x==max(list(map(int, move_to_seoul_zip.keys()))) else x)

# 딕셔너리의 키 값을 조회해서 해당 지역을 출력함
print(f"1970~2016년 동안 제주로 제일 많이 간 사람들의 지역 : {move_to_seoul_zip[str(first_max)]}")
```
