```python
#2번 문제-25줄
import pandas as pd # 판다스 모듈 준비
```

```python
move=pd.read_excel("시도별 전출입 인구수.xlsx") # 엑셀 데이터 읽어들이기
```

```python
data = move[[f"{str(num)}" for num in range(1970,2018)]] # 년도별 에너지 수치 데이터만 따로 저장 ["1970","1971",...,"2016","2017"]
data=data.drop([0])
print(data["1970"].value_counts().count())
for col in data.columns:  # 컬럼명을 반복하면서
    data[col].replace('-','0',inplace=True)
```

```python
for col in data.columns:
    data[col]=data[col].astype("int64") # 수치 데이터로 모두 변환을 했으니 int형으로 Series 형변환시킴
data=data.T # 진우님 아이디어
result=list(data.sum()) # 각 가로줄 별로 합산함 -> 지역에서 지역별로 1970~2017년 동안 이동한 인구의 총 합계를 계산함
```

```python
move=pd.read_excel("시도별 전출입 인구수.xlsx", usecols=['전출지별','전입지별']) # 파일을 새로 불러들임(근데 생각해보니 필요없었음)
```

```python
move = move.drop([0]) # result(합산결과) 와 길이를 맞추기 위해서 move의 데이터 한 행 삭제
move["총 이동인구 수"]=result # "총 이동인구 수" 열을 생성하고 합산 결과를 집어넣음
move
```

```python
local=list(move["전입지별"].unique()) # 현길님 아이디어
#local=["전국","서울","부산","대구","인천","광주","대전","울산","세종특별자치시","경기도","강원도","충청북도","충청남도","전라북도","전라남도","경상북도","경상남도","제주특별자치도"] # 총 합산 결과의 부분과 같이 합쳐져서 Dict로 만들기 위해 지역 리스트 생성
len(local) # 지역의 종류는 총 18가지임
```

```python
default=2 # x 번째로 많이 왔던 지역을 지정(0 : 전국,, 1 : 대부분 서울)
print(f"1970~2016년 동안 각 지역으로 제일 많이 간 사람들의 {default}번째 지역 소개 ") # 소개문구

for i,v in enumerate(local): #이동한 지역별 사람들 수를 comprehension으로 만들기
    move_to=[f'{move["총 이동인구 수"][j+i]}' for j in range(1, 325, 18)] # 각 지역별로 18인덱스씩 건너뛰면서 데이터를 리스트에 수집
    move_to_zip=dict(zip(move_to, local)) # zip으로 local리스트와 move_to데이터를 엮음
    sorted_key=list(map(int,list(move_to_zip.keys()))) # zip으로 합치면서 str로 변경된 키값들을 int형으로 형변환
    sorted_key.sort(reverse=True) # 내림차순 정렬 수행
    print(f"{v:13} <--- {move_to_zip[str(sorted_key[default])]}") # 화살표가 가르키는 방향으로 어떤 지역 사람들이 많이 갔다 출력
```
