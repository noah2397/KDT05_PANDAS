[문1] 아래 조건을 만족하는 Series를 생성하세요.
1) 데이터 : 1~50 범위에서 7의 배수 정수
2) 인덱스 : 'a~z' 범위에서 데이터 갯수 만큼의 알파벳
3) 타  입 : 부호가 없는 1바이트 정수
4) 이  름 : sevens
5) 속  성 : index, values, dtype, ndim, shape, size

```python
import pandas as pd # 모듈 불러오기
import numpy as np #모듈 불러오기...사용은 안했습니다
```

```python
data_list=[i for i in range(7,51,7)] # 데이터 리스트 생성
index_list=[chr(i) for i in range(97, 97+len(data_list))] # 인덱스 리스트 생성
sevens=pd.Series(data=data_list, index=index_list, name="sevens") # 시리즈로 만들기
sevens=sevens.astype("uint8") # 부호가 없는 1바이트 -> 8비트 Unsigned int 로 타입 변경
print(f'info : {sevens.info()}\n')
print(f'index : {sevens.index}') # 각각의 속성 출력
print(f'values : {sevens.values}')
print(f'dtype : {sevens.dtype}')
print(f'ndim : {sevens.ndim}')
print(f'shape : {sevens.shape}')
print(f'size : {sevens.size}')
```

[문3] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stats_100701.xlsx
2) 년도별 인구가 가장 많은 지역과 가장 적은 지역 
3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 
4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 

```python
people=pd.read_excel("stats_100701.xlsx")
people.head()
dataset_local=people.T.iloc[0,2:].to_list() # 지역 이름 리스트를 저장
#2 년도별 인구가 가장 많은 지역과 가장 적은 지역
dataset=people.T.iloc[1:,2:] # 뒤집은 다음 숫자 DataFrame만을 추출
for i in range(4):
    dataset_sum=dataset.iloc[i].to_list() # 년도별 숫자 데이터를 리스트로 저장
    people_dict=dict(zip(dataset_sum,dataset_local)) # 딕셔너리 형태로 묶음
    result=list(people_dict.keys()) # 키를 따로 갖고와서
    print(f'{i+2015}년도별 인구가 가장 많은 지역 : {people_dict[max(result)]}, 가장 적은 지역 : {people_dict[min(result)]}') # max,min으로 출력
```

```python
#3 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역
for i in range(4):
    dataset_sum=dataset.iloc[i].to_list() # 년도별 숫자 데이터를 리스트로 저장
    people_dict=dict(zip(dataset_sum,dataset_local)) # 딕셔너리 형태로 묶음
    result = list(people_dict.keys())[1:7] # 부산,대구,인천,광주,대전,울산
    print(f'{i+2015}년도별 인구가 가장 많은 지역 : {people_dict[max(result)]}, 가장 적은 지역 : {people_dict[min(result)]}') # max,min으로 출력
```

```python
#4 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 
for i in range(4):
    dataset_sum=dataset.iloc[i].to_list() # 년도별 숫자 데이터를 리스트로 저장
    people_dict=dict(zip(dataset_sum,dataset_local)) # 딕셔너리 형태로 묶음
    result = list(people_dict.keys())[8:9]+list(people_dict.keys())[10:16] # 경기, 충북, 충남, 전북, 전남, 경북, 경남
    print(f'{i+2015}년도별 인구가 가장 많은 지역 : {people_dict[max(result)]}, 가장 적은 지역 : {people_dict[min(result)]}') # max,min으로 출력
```

[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
stock=pd.read_excel("stock price.xlsx", index_col="id")
stock.head()
```

```python
#3 (가) 결측치 처리
# 1개라도 NaN이면 모두 제거
stock.dropna()
```

```python
# price가 NaN이면 제거
stock.dropna(subset=["price"])
```

```python
# value가 10000이하면 제거
stock.dropna(subset=["value"])[stock["value"]>10000]
```

```python
#3 (나) 평균값으로 결측치 대체
stock["value"]= stock["value"].fillna(stock["value"].mean())
stock["price"]=stock["price"].fillna(stock["price"].mean())
stock
```

```python
# 이후 값으로 결측치 대체
stock.fillna(method="bfill")
```

```python
# 이전 값으로 결측치 대체
stock.fillna(method="ffill")
```

```python
#4) value가 1000이하 이거나 price가 90000이상인 데이터

filter=(stock["value"]<=1000) | (stock["price"]>=90000)
stock[filter]
```
