```python
video_game_sales = pd.read_csv("../DATA/video_game_sales.csv")
video_game_sales.head()
```

```python
video_game_sales.head(1)
```

```python
video_game_sales.melt(id_vars = 'Name', value_vars = "NA").head()
```

```python
regional_sales_columns = ["NA", "EU", "JP", "Other"]

video_game_sales.melt(
    id_vars = "Name", value_vars = regional_sales_columns
)
```

```python
video_game_sales_by_region = video_game_sales.melt(
    id_vars = "Name", 
    value_vars = regional_sales_columns,
    var_name = "Region",
    value_name = "Sales"
)

video_game_sales_by_region.head()
```

```python
video_game_sales_by_region.pivot_table(
    index = "Name", values = "Sales", aggfunc = "sum", margins=True
).head()
```

```python
cars=pd.read_csv("../DATA/used_cars.csv")
min_wage=pd.read_csv("../DATA/minimum_wage.csv")
print(cars.head(1))
print(min_wage.head(1))
```

```python
#1 cars에서 가격의 합계를 집계하세요. 행 축에서 연료 유형별로 결과를 그룹화하세요
cars.pivot_table(index="Fuel",values="Price", aggfunc="sum")
```

```python
#2 cars에서 자동차의 개수를 집계하세요. 인덱스 축에서 제조업체별로 결과를 그룹화하고 열 축에서 변속기 유형을 그룹화하세요. 행과 열 모두에 대한 소계를 표시하세요
cars.pivot_table(aggfunc="count", index="Manufacturer", columns="Transmission", margins=True).tail() # value 지정 못 함
```

```python
#답지
cars.pivot_table(
    values="Price",
    index="Manufacturer",
    columns="Transmission",
    aggfunc="count",
    margins=True
).tail()
```

```python
#3 cars에서 자동차의 평균 가격을 집계하세요. 인덱스 축에 연도 및 연료 유형별로 결과를 그룹화하고 열 축에 변속기 유형을 그룹화하세요
three=cars.pivot_table(values="Price", aggfunc="mean", index=["Year","Fuel"], columns="Transmission")
three
```

```python
#4 3번 문제의 DataFrame이 주어졌을 때 열 축에서 행 축으로 변속기 유형을 옮기세요
three.stack()
```

```python
#5 min_wage를 넓은 형식에서 좁은 형식으로 변환하세요. 즉, 8개의 연도 열(2010~2017)에 있는 데이터를 하나의 열로 옮기세요
min_wage.melt(id_vars="State", value_vars=[str(i) for i in range(2010,2018)])
```

```python
#답지
min_wage.melt(id_vars="State") # value_vars 매개변수 없이 melt메서드를 호출해도 이전과 동일한 DataFrame을 얻을 수 있음
```

```python
#답지
min_wage.melt(id_vars="State", var_name="Year",value_name="Wage") # value_vars 매개변수 없이 melt메서드를 호출해도 이전과 동일한 DataFrame을 얻을 수 있음
```
