```python
import numpy as np      # nan : Not a Number, 빈칸 의미
from math import nan

data={'col1':['A','A','B',nan,'D','C'],
      'col2':[2,1,9,8,7,4],
      'col3':[0,1,9,4,2,3],
      'col4':['a','B','c','D','e','F'],}

dataDF=pd.DataFrame(data)
dataDF
```

- col 컬럼을 기준으로 오름차순 정렬

```python
# NaN 즉, 빈칸 데이터 (결측치) 값의 위치 : (기본) 제일 마지막
dataDF.sort_values(by='col1')
```

```python
# NaN 즉, 빈칸 데이터 (결측치) 값의 위치 : (기본) 제일 마지막
dataDF.sort_values(by='col1', na_position="first")
```

```python
#col4 컬럼을 기준으로 오름차순 정렬
dataDF.sort_values(by="col4", key=lambda x:x.str.upper())
```
