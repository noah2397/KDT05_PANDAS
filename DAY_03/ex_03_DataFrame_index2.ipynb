```python
import pandas as pd
exam = {'이름' : [ '마징가', '배트맨', '홍길동','임꺽정'],
'국어' : [ 98, 79, 99, 100],
'미술' : [ 70, 99, 87, 80],
'수학' : [ 77, 70, 98, 72],
'체육' : [ 77, 93, 69, 91]}
examDF=pd.DataFrame(exam)
examDF
```

- 행 데이터 추가

```python
# 마징가 98 50 60 77
examDF.loc[4]=['마징가',98,50,60,77]
examDF
```

```python
# 마징가 98 87 77 91
examDF.loc[5]=['마징가', 98,87,70,91]
examDF
```

(5) 새로운 인덱스 설정 - 방법3 : 멀티인덱스 제어 DF.set_index([컬럼1, 컬럼2,..., 컬럼n])

```python
examDF1=examDF.set_index(["이름","국어","수학"])
examDF1
```

- 행 데이터 선택 : DF객체변수명.loc[행레이블] / DF객체변수명.iloc[위치인덱스]

```python
examDF1.loc[('마징가',98,77)]
```

```python
examDF1.loc[('마징가',98,70)]
```

```python
examDF1.loc[('마징가',98)]
```

- 멀티인덱스의 인덱스의 일부분을 기준으로 행 선택 : xs()
    * 예) '마징가' 98 77 => Level 0 1 2

```python
examDF1.xs('마징가') # examDF1.xs('마징가', level=0) 과 똑같다
```

```python
examDF1.loc['마징가']
```

```python
examDF1.xs(98, level=1)
```

```python
examDF1.xs(70, level=2)
```

- ('마징가', 98, 77), ('임꺽정',100,72) 2개 행을 한꺼번에 가져오기

```python
examDF1.loc[[('마징가',98,77), ('임꺽정',100,72)]]
```

```python
examDF1.loc[[("마징가",98,77)]]
```

```python
examDF=examDF.reindex(index=[0,2,4,1,3,5],columns=["이름", "체육","미술","국어","수학","총합"], fill_value=0)
examDF
```
